```python
import time
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import pandas as pd
import seaborn as sns
import scipy.stats as stats
from scipy.stats import poisson

import pylab
from sklearn.preprocessing import StandardScaler

```

```python
def delete_outliers(X_csv, y_csv):
    
    samples, features = X_csv.shape

    del_outliers = []
    for n in range(features):
    
        data_cut =X_csv[n]

        outliers = []


    # 箱行分位差
#     Q3 = data_cut.quantile(q=0.75)
#     Q1 = data_cut.quantile(q=0.25)
#     IQR = Q3-Q1
    
#     upper = Q3 + 1.5*IQR
#     lower = Q1  - 1.5*IQR
    
#     3 sigma 数学分布
        mean= data_cut.mean()
        std = data_cut.std()

        upper = mean + 8*std
        lower = mean  - 8*std


        for i,m  in enumerate(data_cut.index):
            if data_cut[i]< lower or data_cut[i]>upper:
                outliers.append(i)
        del_outliers = del_outliers + outliers
    
    del_outliers = list(set(del_outliers))
    
    X_train_1 = X_csv.drop(X_csv.index[del_outliers]).reset_index(drop=True)
    y_train_1 = y_csv.drop(y_csv.index[del_outliers]).reset_index(drop=True)
    
    return X_train_1, y_train_1
    
    
```

```python
def del_discrete_points(X_csv, y_csv):
    deleted_index = []
    samples, features= X_csv.shape
    for p in range(1, 7):
        data_cut = X_csv[y_csv[0]==p]
        del_index = []
        for n in range(features):
            # 3 sigma
            mean = data_cut.iloc[:, n].mean()
            std = data_cut.iloc[:, n].std()
            upper = mean + 5 * std
            lower = mean - 4 * std
            
            for i, m in enumerate(data_cut.index):
                if (data_cut.iloc[i, n] > upper) or (data_cut.iloc[i, n] < lower):
                    del_index.append(m)
            
        deleted_index = deleted_index + del_index
        deleted_index = list(sorted(set(deleted_index)))
        
    X_train_1 = X_csv.drop(X_csv.index[deleted_index]).reset_index(drop=True)
    y_train_1 = y_csv.drop(y_csv.index[deleted_index]).reset_index(drop=True)
    
    return X_train_1, y_train_1
```

```python
def cut_dim_by_2_features_corr(X_tain_clean, y_tain_clean):
    delete_columns_dist = {}
    n = X_tain_clean.shape[1]
    
    for i in range(n):
        delete_columns_dist[i] = []
        for j in range(i + 1, n):
            corr_i_j = stats.pearsonr(X_tain_clean.iloc[:, i], X_tain_clean.iloc[:, j])
            if corr_i_j[0] >= 0.96:
                delete_columns_dist[i].append(j)
    
    del_columns = []
    for key, values in delete_columns_dist.items():
        del_columns = del_columns + values
        
    del_columns = list(set(del_columns))
    return del_columns
```

```python
def cut_dim_by_features_corr_label(X_csv, y_csv, del_1st_dims):
    
    X_del_dims_1st = X_csv.drop(columns=X_csv.columns[del_1st_dims])
    
    X_re_fes = pd.DataFrame(X_del_dims_1st)

    X_re_fes['label'] = y_csv
    
    index_corrScore_xy = {}

    # Compute Pearson Correlation between label and each feature,
    for i, true_index in enumerate(X_re_fes.columns):
        
        if i == len(X_re_fes.columns) - 1:
            break
            
        a = X_re_fes.iloc[:, i]
        b = X_re_fes.iloc[:, -1]
        
        corr_ab = stats.pearsonr(a, b)[0]
        corr_ab = round(corr_ab, 7)
        
        index_corrScore_xy[true_index] = corr_ab

#     output = pd.DataFrame(
#         {'True_Index': list(index_corrScore_xy.keys()), 'feature_corr_label': list(index_corrScore_xy.values())})
#     output.sort_values(by="feature_corr_label", ascending=False, inplace=True)

    del_2nd_index = []
    
    for key, value in index_corrScore_xy.items():
        if - 0.15 < index_corrScore_xy[key] < 0.15:
#             del_2nd_index.append(key)
#         if 0 < index_corrScore_xy[key] < 0.15:
            del_2nd_index.append(key)
    
    del_2nd_index = list(set(del_2nd_index))
    return del_2nd_index
```

```python
def filter_features(X_train_1, y_train_1):
    

    del_1st_dims = cut_dim_by_2_features_corr(X_train_1, y_train_1)
    del_2nd_dims = cut_dim_by_features_corr_label(X_train_1, y_train_1, del_1st_dims)

    del_dims = del_1st_dims + del_2nd_dims
    del_dims = [int (i) for i in del_dims]
    
    return del_dims


def del_features(X, y, del_features):
    X = X.drop(columns = X.columns[del_features]).T.reset_index(drop=True).T

    
    
#     X= X.values
#     y = y.values.ravel()
    
    return X, y
```

```python
from sklearn.preprocessing import StandardScaler

print("Start : %s\n" % time.ctime())

X_train_csv = pd.read_csv("./MLinTheUnknown-Data/X_train.csv", header=None)
y_train_label_csv = pd.read_csv("./MLinTheUnknown-Data/y_train.csv", header=None)

X_train_kmeans  = pd.read_csv("../project/X_train_Kmeans.csv")
y_train_kmeans  = pd.read_csv("../project/y_train_Kmeans.csv")


X_val_csv = pd.read_csv("./MLinTheUnknown-Data/X_val.csv", header=None)
y_val_label_csv = pd.read_csv("./MLinTheUnknown-Data/y_val.csv", header=None)


### Step1, ####feature selection, including training sets and validation sets

del_dims = filter_features(X_train_csv, y_train_label_csv)

X_train, y_train = del_features(X_train_csv,y_train_label_csv, del_dims)
X_val,  y_val  = del_features(X_val_csv,y_val_label_csv, del_dims)



# step2, del the outliers noise

# X_train, y_train = del_discrete_points(X_train, y_train)
X_train, y_train = delete_outliers(X_train, y_train)



from sklearn.decomposition import PCA, IncrementalPCA
from sklearn.pipeline import make_pipeline

pca = PCA(n_components=11)


X_train = pca.fit_transform (X_train_csv)
X_val = pca.transform (X_val_csv)



# # # step3, normalization 
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

y_train = y_train.values.ravel()


X_val = scaler.transform(X_val)
y_val = y_val.values.ravel()


# X_train = X_train_csv.values
# y_train = y_train_label_csv.values.ravel()

# X_val = X_val_csv.values
# y_val = y_val_label_csv.values.ravel()



print("End : %s" % time.ctime())
```
